````python
def plotImages(lstFiles):
    _, axs = plt.subplots(1, 4, figsize=(20, 20))
    axs = axs.flatten()
    for img, ax in zip(lstFiles, axs):
        rgb_image = Image.open(img)
        #print("dkdk")
        ax.imshow(rgb_image)
    plt.show()
````

````python
lstBoots=["test_images/boots1.jpg","test_images/boots2.jpg","test_images/boots3.jpg","test_images/boots4.jpg"]
lstVests=["test_images/vest1.jpg","test_images/vest2.jpg","test_images/vest3.jpg","test_images/vest4.jpg"]
lstMasks=["test_images/mask1.jpg","test_images/mask2.jpg","test_images/mask3.jpg","test_images/mask4.jpg"]
lstHats=["test_images/hat1.jpg","test_images/hat2.jpg","test_images/hat3.jpg","test_images/hat4.jpg"]
print("Plotting Boots")
plotImages(lstBoots)
print("Plotting Vests")
plotImages(lstVests)
print("Plotting Masks")
plotImages(lstMasks)
print("Plotting Hats")
plotImages(lstHats)

````
